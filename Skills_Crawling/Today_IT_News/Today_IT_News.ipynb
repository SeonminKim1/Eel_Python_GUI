```python
# Crawling Python
# Today IT News
# (1) 오늘 날짜 IT 기사들 제목 가져오기
# (2) csv 파일로 저장하기

from bs4 import BeautifulSoup
import urllib.request as req
from datetime import datetime # 날짜관련
import csv, codecs

# IT NEWS 가져올 페이지 수
pages = 10

# 기사 제목 리스트
title = []
url_forward = "https://news.naver.com/main/list.nhn?mode=LS2D&mid=shm&sid2=230&sid1=105&date="
for page in range(1, pages):
    # 오늘 날짜 YYYYmmddHHMMSS 형태의 시간 출력 url 중간부분
    url_mid = datetime.today().strftime("%Y%m%d") 
    url_end = "&page=" + str(page) # pages 들
    
    # 최종 우리가 가져올 url (페이지마다)
    final_url = url_forward + url_mid + url_end

    # url request 요청
    res = req.urlopen(final_url)
    soup = BeautifulSoup(res, 'html.parser')
    
    # 리스트형으로 나옴. select라서
    itnews_day = soup.select("#main_content > div.list_body.newsflash_body > ul.type06_headline > li > dl > dt > a") 
    
    # 뽑은 제목들 리스트 만들기
    for sentence in itnews_day:
        title.append(str.strip(sentence.text)) 
        
# 리스트 내 공백 지우는 방법
c_lst = [x for x in title if x]
title = c_lst

# 리스트 한줄씩 출력
for i, t in enumerate(title):
    print(i+1 , t)
    
# CSV 파일로 저장하기
Filename = str(datetime.today().strftime("%Y%m%d")) +'_IT_News.csv'
with codecs.open(Filename, "w", "utf-8") as fp:
    writer = csv.writer(fp, delimiter = ",", quotechar='"')
    writer.writerow(["번호", "기사제목"])
    for i, name in enumerate(title):
        writer.writerow([i+1, name])
print('\n\n' + Filename + " 저장완료")
```

### 서울시 공공데이터 지하철 역별 승하차 인원 (190326)
- JSON 파일로 제공해주는 지하철 역별 승하차 인원 JSON 파일을 파싱하여 호선별 CSV파일로 저장

### 변수명
- "USE_DT":"사용일자","SUB_STA_NM":"역명","WORK_DT":"등록일자",
- "RIDE_PASGR_NUM":"승차총승객수","LINE_NUM":"호선명","ALIGHT_PASGR_NUM":"하차총승객수"

```python
import json
import csv, codecs 

# 1. json 파일 읽어오기.
items = json.load(open("Information on boarding and departing passengers by Seoul subway line(190326).json", "r", encoding="utf-8"))

day = "20190326" # 측정할 날짜 (use_dt)
station_name = ["1호선","2호선","3호선","4호선","5호선","6호선","7호선","8호선","9호선",
                "9호선2~3단계","경강선","경부선","경원선","경의선","경인선","경춘선",
                "공항철도 1호선","과천선","분당선","수인선","안산선","우이신설선","일산선",
                "장항선","중앙선"]
print("지하철 호선 갯수는 " , str(len(station_name)), '개')
```

```python
# 2, csv 쓰기 모드 열기 (with)
for i in station_name:
    count=0 
    with codecs.open("Subway " + i + ".csv", "w", "euc_kr") as fp:
        writer = csv.writer(fp, delimiter=",", quotechar='"') # writer 형식      
        writer.writerow(["날짜", i, "역이름","타는인원","내린 인원"]) # 첫 행 (열 항목 이름들)
        
        # 3. json 파일 출력하기
        for station in items["DATA"]: # 처음형태가 dict 이라서 DESCRIPTION 항목과 DATA항목중 DATA로 들어감.

            #print(station["line_num"]) # 모든 역에 대한 이름
            if station["line_num"] == i and station["use_dt"] == day:
                count = count + 1

                # 4. csv에 입력 
                writer.writerow([day, # 날짜
                                 i, # 호선 이름
                                 station["sub_sta_nm"], # 역이름
                                 station["ride_pasgr_num"], # 타는 인원
                                 station["alight_pasgr_num"]])  # 내리는 인원

                # 5. 콘솔 출력
                print(day, "에", station["sub_sta_nm"], "역에서",
                      "타는 인원은 " , station["ride_pasgr_num"],
                      "내리는 인원은 ", station["alight_pasgr_num"])
        print(i + '역 갯수는', str(count))

```

## Scatter Plot
- **두 feature간의 관계**를 알기 위해 사용하는 그래프
    - 상관 관계 확인 (양의 상관관계 / 음의 상관관계 / 없음)
- **군집 / 값 그룹 간의 차이 / 이상치 등 도 확인**하기 좋음
- 점이 많아질수록 점의 분포를 파악하기 힘들다! (해결방법)
    - 투명도 조정
    - 지터링 (jittering) : 점의 위치를 약간씩 변경
    - 2차원 히스토그램 : 히트맵을 사용하여 깔끔한 시각화
    - Contour plot : 분포를 등고선을 사용하여 표현
    
- Seaborn이 좀더 잘 되있음.

### 1. Scatter 기본

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
print('Matplotlib Version : ', mpl.__version__)
```

```python
fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111, aspect=1) # aspect : 가로축과 세로축의 scale맞춰줌

x, y = np.random.rand(20), np.random.rand(20)
ax.scatter(x,y)
ax.set_xlim(0, 1.05)

plt.show()
```

### 2. Scatter Plot 요소
- 색(color)
- 모양(marker)
- 크기(size)

```python
fig = plt.figure(figsize=(7,7))
x, y = np.random.rand(20), np.random.rand(20)

ax1 = fig.add_subplot(121, aspect=1) # aspect : 가로축과 세로축의 scale맞춰줌
ax1.scatter(x,y,
          s= 150, # 점의갯수
          c = 'royalblue', # oolor
          marker='o',)

# marker크기 다양하게
s = np.arange(20) * 20
ax2 = fig.add_subplot(122, aspect=1) # aspect : 가로축과 세로축의 scale맞춰줌
ax2.scatter(x, y,
          s= s, # 점의갯수
          c = 'royalblue', # oolor
          marker='o',)

plt.show()
```
